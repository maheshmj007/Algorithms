```python
import numpy as np
import sys
# enter adjacency matrix as a array graph
graph = np.array([[0, 4, 0, 0, 0, 0, 0, 8, 0], 
        [4, 0, 8, 0, 0, 0, 0, 11, 0], 
        [0, 8, 0, 7, 0, 4, 0, 0, 2], 
        [0, 0, 7, 0, 9, 14, 0, 0, 0], 
        [0, 0, 0, 9, 0, 10, 0, 0, 0], 
        [0, 0, 4, 14, 10, 0, 2, 0, 0], 
        [0, 0, 0, 0, 0, 2, 0, 1, 6], 
        [8, 11, 0, 0, 0, 0, 1, 0, 7], 
        [0, 0, 2, 0, 0, 0, 6, 7, 0] 
        ])
```

```python
spt_set = set(np.linspace(0,8,num=9,dtype='int'))
dist = dict(zip(spt_set,[sys.maxsize]*9))

def mindist(dist,spt_set):
    min_i = sys.maxsize
    index = -1
    for i in spt_set:
        if dist[i] < min_i:
            index = i
            min_i = dist[i]
    return index    
```

```python
src = 0
dest = 6
dist[src] = 0

while len(spt_set) > 0:
    spt_set.remove(src)
    print('source',src)
    edge = np.nonzero(graph[src])
    for e in edge[0]:
        if e in spt_set and dist[e] > dist[src] + graph[src][e]:
            dist[e] = dist[src] + graph[src][e]
    print(dist)
    src = mindist(dist,spt_set)
print('Dijkstraâ€™s shortest distance :',dist[dest])
```
